```
#TASK 1 (SET UP) + TASK 2 (CHI-SQUARED)

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


def shoot_particle():
    # screen distances is the array containing the distance between each screen and the origin
    n_screens = 8
    screen_distances = np.arange(n_screens)*100+200
    screen_distances

    # max_angle is the maximal degree at which a photon can be shot
    max_angle = 15
    #Generate random angle of x and y from -15 to +15
    x_angle = (np.random.rand()*2*max_angle-max_angle)*(np.pi/180)
    y_angle = (np.random.rand()*2*max_angle-max_angle)*(np.pi/180)
    x_angle
    y_angle

    # Find x/y coordinates of hit on the screen
    x_hit = np.tan(x_angle)*screen_distances
    y_hit = np.tan(y_angle)*screen_distances

    # screen is 1000x1000 big and has 80x80 pixels
    screen_size = 1000
    pixels = 80

    #Screen divided from -500 to 0 to +500. Pixels divided from -40 to 0 to +40.
    # finds which pixel is hit

    #Hit_x_pixel is (x_hit /500)*pixels to convert x-coordinate to pixels
    #Hit_y_pixel is (y_hit /500)*pixels to convert y-coordinate to pixels
    hit_x_pixel = (x_hit/(screen_size/2)*pixels).astype(int)
    hit_y_pixel = (y_hit/(screen_size/2)*pixels).astype(int)
    if max(hit_x_pixel) > 40:
        print(x_hit)
        print(hit_x_pixel)
    # with 10% chance move the hit to adjacent pixel
    #Parse through hit_x_pixel and hit_y_pixel indices, if random number generated less than or equal to 0.1 (10%), then move pixel at that index for both x and y pixel by -1, 0 or 1
    for i in range(len(hit_x_pixel)):
        if np.random.rand() <= 0.1:
            # print('test')
            hit_x_pixel[i]+=np.random.randint(3)-1
            hit_y_pixel[i]+=np.random.randint(3)-1
    #Returns 2 arrays each 8 long, for x pixels for each screen and y pixels for each screen. Hit_x_pixel and hit_y_pixel, return values between -40 to 0 to +40. As indexing from 0,0 to 80,80 if hit at 0,0, itâ€™s actually 40,40
    #hence adding 40 to both arrays.
    return np.clip(hit_x_pixel+int(pixels/2),0,pixels-1), np.clip(hit_y_pixel+int(pixels/2), 0,pixels-1)


def get_chi_2(hit_x_pixel, hit_y_pixel):  

    #Z_detection point = array(200, 300, 400, 500, 600, 700, 800, 900)
    z_det = np.arange(n_screens)*100+200

    #X_detection point = array(ratio of hit_x_pixels to total number of pixels* screensize)
    x_det = np.array(hit_x_pixel*screen_size/80-screen_size/2, ndmin=2)

    #Y_detection point = array(ratio of hit_y_pixels to total number of pixels* screensize)
    y_det = np.array(hit_y_pixel*screen_size/80-screen_size/2, ndmin=2)

    print("X_det", x_det)
    print("Y_det", y_det)

    #XY_detection points = array( x_det, y_det) thus (x,y) coordinates
    xy_det = np.concatenate((x_det, y_det), axis=0)

    # plot actual measured locations for x and y vs z (distance to screen)
    plt.axes(xlabel = 'Z detection points', ylabel = 'x and y detection points')
    plt.plot(z_det, x_det.T, color='blue')
    plt.plot(z_det, y_det.T, color='green')

 
    #Obtain polynomial coefficients for 1st degree polynomial, so for ax + b, returns a and b
    a, b = np.polyfit(z_det, xy_det.T, 1)

    # Preds (predictions) for xy coordinates = ax + b, where a = 2D array of a, x = z_detection_points, b = 2D array of b.
    preds = (np.array(a,ndmin=2).T*np.array(z_det,ndmin=2)+np.array(b,ndmin=2).T).T

    # plot best fit
    plt.plot(z_det, preds, color='orange')
    plt.show()
    
    #Variance = width of a pixel in coordinates
    sigma_2 = (screen_size/pixels)**2

    print(sigma_2)

    # get chi squared value
    sum = 0
    for i in range(len(preds)):
        sum+=(xy_det.T[i]-preds[i])**2/sigma_2
        print(sum)
    return sum



screen_size = 1000
pixels = 80
n_screens = 8
screens = np.zeros((n_screens, pixels, pixels))

chi_2_arr = []
#For 10 particles: hit_x_pixel array and hit_y_pixel retrieved from shoot_particle function. 

#Then, for each particle, append Chi 2 value retrieved from get_chi_square, for hit_x_pixel and hit_y_pixel method. 
#To the screens array, increment x pixel and y pixel by 1.


for i in range(10):
    hit_x_pixel, hit_y_pixel = shoot_particle()
    print('hit x pixel', hit_x_pixel)
    print('hit y pixel', hit_y_pixel)
    chi_2_arr.append(get_chi_2(hit_x_pixel, hit_y_pixel))
    screens[:,hit_x_pixel, hit_y_pixel]+=1

#X_chi2 = array of 0th index values of chi_2_array
x_chi2 = np.array(chi_2_arr).T[0]
print('chi 2 values for x', x_chi2)
#Y_chi2 = array of 1st index values of chi_2_array
y_chi2 = np.array(chi_2_arr).T[1]
print('chi 2 values for y', y_chi2)
plt.axes(xlabel = 'Chi square value for x coordinates', ylabel = 'Frequency')
plt.hist(x_chi2, alpha=0.5, color='blue', bins = 25)

plt.show()

plt.axes(xlabel = 'Chi square value for y coordinates', ylabel = 'Frequency')
plt.hist(y_chi2, alpha=0.5, color='orange', bins = 25)

plt.show()
```
