```
#atomic decay
import random
import matplotlib.pyplot as plt
import math

#can change to 1000/10000
data_xaxis = [random.randint(1,10) for i in range(10)]

#create new array for data for xi on y axis
data_yaxis=[]

for i in range(10):
    data_yaxis.append(data_xaxis[i-1])

#print statemnts for seeing data

print(data_xaxis)
print(data_yaxis)

#plot graph
#new run = new graph everytime, indicating randomness

plt.plot(data_xaxis, data_yaxis)
plt.show()
```

```
#PROBABILITY OF DECAY

#number of undecayed atoms = 5000
atoms_sample = 5000
#halflife = 500 seconds
t_halflife = 500
#list of decayed atoms with 0 at every time step
decayed_atoms = [0 for i in range(1000)]
#list of undecayed atoms with 5000 atoms at every time step
atoms_left = [5000 for i in range(1000)]

#probability function for calculating prob of decay
def probability (time):
 prob_of_decay = 1-(2 ** ((0-time)/t_halflife))
 return prob_of_decay

#def number_atoms (time):
  #atoms_number = atoms_sample*(2**((0-time)/t_halflife))
  #return atoms_number
  
#for each time step
for t in range(1000):
  for n in range(atoms_sample):
#random probability between 0 and 1
    prob = random.random()
#probability of decay at each time step - calling probability function
    prob_of_decay = probability(t)
    if prob_of_decay > prob:
      atoms_sample -= 1
      decayed_atoms[t] += 1
    if t > 0:
      atoms_left[t] = atoms_left[t-1] - decayed_atoms[t]

decayed_atoms
#graph of decayed atoms
plt.plot(decayed_atoms[0:200])

```
