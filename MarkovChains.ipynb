```
#Markov Chains

import random
import numpy as np
import matplotlib.pyplot as plt

weather = ['sunny','rainy']
days = 14
numSims = 50
#Can change probabilities here, will change overall transition_matrix
#P(A|A)
prob_sungiven_sun = 0.9
#P(B|A)
prob_raingiven_sun = 0.1
#P(A|B)
prob_sungiven_rain = 0.5
#P(B|B)
prob_raingiven_rain = 0.5

# [[P(A|A),P(B|A)],[P(A|B),P(B|B)]]
transition_matrix = [[prob_sungiven_sun,prob_raingiven_sun], [prob_sungiven_rain,prob_raingiven_rain]]
print(transition_matrix)

transition_names = [['A_A' , 'B_A'], ['A_B', 'B_B']]


def Forecast(no_of_days):


    #choose weather first day
    weatherFirstDay = random.choice(weather)
    d = 0
    #new list for weather for each day
    Forecast.weather_forc = []
    print("Weather on first day: ",weatherFirstDay)
    #create forecast
    while d < no_of_days:
        if weatherFirstDay == "sunny":
            #numpy.random.choice(a, size=None, replace=True, p=None)
            tomorrow = np.random.choice(transition_names[0],replace=True,p=transition_matrix[0])
            Forecast.weather_forc.append(weatherFirstDay)
            
            if tomorrow == "A_A":
                pass
            else:
                weatherFirstDay = "rainy"
        elif weatherFirstDay == "rainy":
            tomorrow = np.random.choice(transition_names[1],replace=True,p=transition_matrix[1])
            Forecast.weather_forc.append(weatherFirstDay)
            if tomorrow == "B_B":
                pass
            else:
                weatherFirstDay = "sunny"
       
        d += 1
        
    return Forecast.weather_forc  

         

# We forecast the weather for 14 days
#print (Forecast(days))

#no_of_sunny = Forecast.weather_forc.count('sunny')
#no_of_rainy = Forecast.weather_forc.count('rainy')

#print (" no of sunny days: ", no_of_sunny )
#print (" no of rainy days " , no_of_rainy)

i = 0
#lists for counting occurences of sunny and rainy in simulations
sunny_count = []
rainy_count = []

#50 simulations
for i in range (numSims):
  
  print ("The forecast for" , i , "simulation is :" , Forecast(days))
  no_of_sunny = Forecast.weather_forc.count('sunny')
  sunny_count.append(no_of_sunny)
  no_of_rainy = Forecast.weather_forc.count('rainy')
  rainy_count.append(no_of_rainy)
  i += 1
print (sunny_count)
print (rainy_count)  

#print (" no of sunny days: ", no_of_sunny )
#print (" no of rainy days " , no_of_rainy)

#graph simulation
numSims = list(range(1, (numSims + 1)))
print(list(range(14)))
plt.bar(numSims, rainy_count, label="Rainy Days")
plt.bar(numSims, sunny_count, bottom= rainy_count, color='gold', label="Sunny Days")
plt.ylabel("Days")
plt.xlabel("Simulation Number")
plt.legend(loc="upper right")
plt.title("Simulated Weather Forecast over 14 days")
plt.show()
```
